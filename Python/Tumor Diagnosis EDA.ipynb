```python
# Importing libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
df = pd.read_csv('C:\\Users\\KIIT\\Downloads\\data.csv')
df
```

```python
df.columns
```

```python
y= df.diagnosis
dropped_col= {'id', 'diagnosis', 'Unnamed: 32'}
x= df.drop(dropped_col, axis=1)
x.head()
```

```python
ax= sns.countplot(y, label="Count")
B, M= y.value_counts()
print("Number of Benign tumors:", B)
print("Number of Malignant tumors:", M)
```

```python
x.describe()
```

```python
# Violinplot

df_x= x
df_x_std= (df_x - df_x.mean()) / df_x.std()
df_x= pd.concat([y, df_x_std.iloc[:, 0:10]], axis=1)
df_x= pd.melt(df_x, id_vars='diagnosis', var_name='Features', value_name='Value')   # pandas.melt- Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.
plt.figure(figsize=(10,10))
sns.violinplot(x= 'Features', y= 'Value', hue='diagnosis', data=df_x, split= True, inner= 'quartile')
plt.xticks(rotation=45)
```

```python
df_x= pd.concat([y, df_x_std.iloc[:, 10:20]], axis=1)
df_x= pd.melt(df_x, id_vars='diagnosis', var_name='Features', value_name='Value')   # pandas.melt- Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.
plt.figure(figsize=(10,10))
sns.violinplot(x= 'Features', y= 'Value', hue='diagnosis', data=df_x, split= True, inner= 'quartile')
plt.xticks(rotation=45)
```

```python
# Boxplot

plt.figure(figsize=(10,10))
sns.boxplot(x= 'Features', y= 'Value', hue= 'diagnosis', data= df_x)
plt.xticks(rotation= 45)
```

```python
# Jointplot

sns.jointplot(x.loc[:, 'concavity_worst'], x.loc[:, 'concave points_worst'], kind= 'reg')
```

```python
# Swarmplot

df_x= pd.concat([y, df_x_std.iloc[:, 0:10]], axis=1)
df_x= pd.melt(df_x, id_vars='diagnosis', var_name='Features', value_name='Value')
plt.figure(figsize=(30,30))
sns.swarmplot(x= 'Features', y= 'Value', hue='diagnosis', data=df_x, size= 5)
plt.xticks(rotation=45)
```
